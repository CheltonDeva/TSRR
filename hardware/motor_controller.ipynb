```python
import RPi.GPIO as GPIO
import time

class MotorController:
    def __init__(self):
        # Motor A (Left) pins
        self.LEFT_ENABLE = 25
        self.LEFT_FWD = 24
        self.LEFT_BWD = 23
        
        # Motor B (Right) pins
        self.RIGHT_ENABLE = 17
        self.RIGHT_FWD = 27
        self.RIGHT_BWD = 22
        
        self.setup_gpio()
    
    def setup_gpio(self):
        """Initialize GPIO pins."""
        GPIO.setmode(GPIO.BCM)
        GPIO.setwarnings(False)
        
        # Setup all pins
        pins = [
            self.LEFT_ENABLE, self.LEFT_FWD, self.LEFT_BWD,
            self.RIGHT_ENABLE, self.RIGHT_FWD, self.RIGHT_BWD
        ]
        
        for pin in pins:
            GPIO.setup(pin, GPIO.OUT)
            GPIO.output(pin, GPIO.LOW)
        
        # Setup PWM
        self.left_pwm = GPIO.PWM(self.LEFT_ENABLE, 100)
        self.right_pwm = GPIO.PWM(self.RIGHT_ENABLE, 100)
        self.left_pwm.start(0)
        self.right_pwm.start(0)
    
    def move_forward(self, speed=50):
        """Move the vehicle forward."""
        self.left_pwm.ChangeDutyCycle(speed)
        self.right_pwm.ChangeDutyCycle(speed)
        GPIO.output(self.LEFT_FWD, GPIO.HIGH)
        GPIO.output(self.LEFT_BWD, GPIO.LOW)
        GPIO.output(self.RIGHT_FWD, GPIO.HIGH)
        GPIO.output(self.RIGHT_BWD, GPIO.LOW)
    
    def move_backward(self, speed=50):
        """Move the vehicle backward."""
        self.left_pwm.ChangeDutyCycle(speed)
        self.right_pwm.ChangeDutyCycle(speed)
        GPIO.output(self.LEFT_FWD, GPIO.LOW)
        GPIO.output(self.LEFT_BWD, GPIO.HIGH)
        GPIO.output(self.RIGHT_FWD, GPIO.LOW)
        GPIO.output(self.RIGHT_BWD, GPIO.HIGH)
    
    def turn_left(self, speed=50):
        """Turn the vehicle left."""
        self.left_pwm.ChangeDutyCycle(speed)
        self.right_pwm.ChangeDutyCycle(speed)
        GPIO.output(self.LEFT_FWD, GPIO.LOW)
        GPIO.output(self.LEFT_BWD, GPIO.HIGH)
        GPIO.output(self.RIGHT_FWD, GPIO.HIGH)
        GPIO.output(self.RIGHT_BWD, GPIO.LOW)
    
    def turn_right(self, speed=50):
        """Turn the vehicle right."""
        self.left_pwm.ChangeDutyCycle(speed)
        self.right_pwm.ChangeDutyCycle(speed)
        GPIO.output(self.LEFT_FWD, GPIO.HIGH)
        GPIO.output(self.LEFT_BWD, GPIO.LOW)
        GPIO.output(self.RIGHT_FWD, GPIO.LOW)
        GPIO.output(self.RIGHT_BWD, GPIO.HIGH)
    
    def stop(self):
        """Stop the vehicle."""
        self.left_pwm.ChangeDutyCycle(0)
        self.right_pwm.ChangeDutyCycle(0)
        GPIO.output(self.LEFT_FWD, GPIO.LOW)
        GPIO.output(self.LEFT_BWD, GPIO.LOW)
        GPIO.output(self.RIGHT_FWD, GPIO.LOW)
        GPIO.output(self.RIGHT_BWD, GPIO.LOW)
    
    def cleanup(self):
        """Clean up GPIO resources."""
        self.stop()
        GPIO.cleanup()
```
